```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('coffee.csv')
df.head(10)
```

```python
x=df.drop(columns=['Kahve Tipi'])
y=df['Kahve Tipi']
```

```python
scaler = MinMaxScaler()
normalized_x = scaler.fit_transform(x)
```

```python
user_input = {
    'Kahve (ml)': 100,
    'Süt (ml)': 0,
    'Su (ml)': 1000
}

user_df = pd.DataFrame([user_input])
user_df = user_df[['Su (ml)', 'Süt (ml)', 'Kahve (ml)']] # Reorder columns to match the training data
user_normalized = scaler.transform(user_df)
```

```python
similarities = cosine_similarity(user_normalized, normalized_x)
top_indexes = np.argsort(similarities[0])[::-1][:3]
print("En uygun kahve önerileri:")
for idx in top_indexes:
    print(f"- {df.loc[idx, 'Kahve Tipi']} {x.iloc[idx].to_dict()} (Benzerlik: {similarities[0][idx]:.2f})")
```
