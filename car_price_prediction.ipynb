```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('audi.csv')
```

```python
# y= ax+b
```

```python
df.head(3)
```

```python
df=df.drop(columns=['index','href','MileageRank','PriceRank','PPYRank','Score'])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.head(3)
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahip","fiyat","ppy"]
```

```python
df.head(3)
```

```python
df['motor']=df['motor'].str.replace("L","")
```

```python
df['motor']=pd.to_numeric(df['motor'])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True)
```

```python
y=df['fiyat']
x=df.drop("fiyat",axis=1)
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```

```python
model.predict([[2017,30000,1.6,110,1,2600,0,1]])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=22)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
